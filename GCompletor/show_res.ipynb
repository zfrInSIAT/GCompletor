```python
import numpy as np
from sklearn import metrics
def mape(y_true, y_pred):
    return np.mean(np.abs(y_pred-y_true)/y_true)*100

def rmse(y_true, y_pred):
    return np.sqrt(metrics.mean_squared_error(y_pred,y_true))
```

```python
# ori_data shape (day,time_slot,road) 
def mean_imputation(ori_data):
    data_shape = ori_data.shape
    mean_ten = np.zeros((data_shape[1],data_shape[2]))
    for i in range(data_shape[2]):
        for j in range(data_shape[1]):
            sum_ = 0
            count = 0
            for k in range(data_shape[0]):
                if ori_data[k,j,i]!=0:
                    sum_+=ori_data[k,j,i]
                    count+=1
            if count!=0:
                mean_ten[j,i]=sum_/count
    return mean_ten
weekday_mean = mean_imputation(weekday_all_data)
weekend_mean = mean_imputation(weekend_all_data)
weekday_mean_ten = np.repeat(np.expand_dims(weekday_mean,axis=0),10,axis=0)
weekend_mean_ten = np.repeat(np.expand_dims(weekend_mean,axis=0),4,axis=0)
```

```python
# weekday
x_list = []
y_rmse_list = []
y_mape_list = []
for i in range(288):
    x_list.append(i)
    rmse_ = rmse(weekday_all_data[:,i][mask_weekday_all_data[:,i]],weekday_mean_ten[:,i][mask_weekday_all_data[:,i]])
    mape_ = mape(weekday_all_data[:,i][mask_weekday_all_data[:,i]],weekday_mean_ten[:,i][mask_weekday_all_data[:,i]])
    y_rmse_list.append(rmse_)
    y_mape_list.append(mape_)
```

```python
print(x_list)
print("weekday_rmse:",y_rmse_list)
print("weekday_mape:",y_mape_list)
```
